### Create graph from adjacency matrix
We can also built the network by input a adjacent matrix.

Firstly, lets compute a adj matrix from the edgelist and visualize it.

```python
adj0 = nx.adjacency_matrix(G,weight='weight')
betterspy.show(adj0)
```

Note that adj0 is of sparse matrix type. Some usual numpy operations on dense matrix may not work. For example, if you want to see and read it, you need to turn into a dense type.

```python
adj0
```

```python
# notice adj0 is a sparse matrix to print it need to turn into a dense type
adj0.todense()
```

 Let's turn this into a unweighted directed matrix. We are going to built network using the new matrix.

```python
unweighted_adj = adj0.copy() # so change in unweighted_adj won't affect adj0
unweighted_adj[unweighted_adj>0]=1
print(unweighted_adj.todense())
plt.figure(figsize=(8,8))
plt.spy(unweighted_adj) # another way to visualize sparse matrix
```

```python
G_unweighted = nx.DiGraph(unweighted_adj)
G_unweighted.edges.data()
```

```python
# Notice we still have 'weight' feature although they all have value 1. We can del the feature.|
for u,v in G_unweighted.edges:
    del G_unweighted[u][v]['weight']
```

```python
nx.is_weighted(G_unweighted)
```

```python
nx.draw(G_unweighted,pos = nx.spring_layout(G_unweighted,k=5),with_labels=True) # Take times for large network
plt.axis("on");
```

- pos parameters allow user to select which node layout (coordinates in plotting )

### Watts strogatz graph

Use the networkx.random_graphs.watts_strogatz_graph to generate some WS graph with parameters you choose. Try to visualize the graph using nx.draw in a suitable layout.

Try out the example in L3 slides. Changing the p from 0-1 inclusive and check how this influence the averge shortest path between vertices and average clustering coefficient. Plot the normalized values of these two quantities. What do you observe.

```python
p_list = np.arange(0,1,0.05)
N = 50;
k = 6;
average_sp = np.zeros((len(p_list),1))# average shortest path list
average_cc = np.zeros((len(p_list),1))# average clustering coef list
for idx,p in enumerate(p_list):

    # Generate Watts strogatz graph using the networkx random graph generator
    G_temp = nx.random_graphs.watts_strogatz_graph(N,k,p)
    # visualize a network using circular layout
    print('p value is ',p)
    nx.draw(G_temp,pos=nx.circular_layout(G_temp),with_labels=True,node_size=200,font_color='white')
    plt.show()
    average_sp[idx] = nx.average_shortest_path_length(G_temp)
    average_cc[idx] = nx.average_clustering(G_temp)
```

```python
plt.figure(figsize=(10,8))
# Normalize starting point to be 1
plt.plot(p_list,average_sp/(average_sp[0]),'o-',label = 'average shortest path')
plt.plot(p_list,average_cc/(average_cc[0]),'o-',label = 'average clustering coef')
plt.legend(fontsize=18)
plt.grid()
plt.ylabel('Normalized value',fontsize=18)
plt.xlabel('p',fontsize=18)
```

Notice, there are some variants regarding how you rewireing the edges in WS graph mentioned in lecture. You can try to code your own WS graph generator and check the impact of p using your own function.
