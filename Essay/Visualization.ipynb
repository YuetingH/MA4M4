### Create Graph

```python
G1 = nx.DiGraph(); G2 = nx.DiGraph()
```

```python
# Add nodes
node_name = node_df[["selector"]].values.tolist(); node_name = sum(node_name, [])
node_polarity = node_df[["polarity"]].values.tolist(); node_polarity = sum(node_polarity, [])

node_pos = node_df[["x", "y"]].values.tolist(); node_pos = [tuple(x) for x in node_pos]

fan_size_t1 = node_df[["fan_size_t1"]].values.tolist(); fan_size_t1 = sum(fan_size_t1, [])
fan_size_t2 = node_df[["fan_size_t2"]].values.tolist(); fan_size_t2 = sum(fan_size_t2, [])

for i in range(len(node_name)):
    G1.add_node(node_name[i], polarity = node_polarity[i], pos = node_pos[i], fan_size_t1 = fan_size_t1[i])
    G2.add_node(node_name[i], polarity = node_polarity[i], pos = node_pos[i], fan_size_t2 = fan_size_t2[i])
```

```python
# Add edges
edge1 = edge_df1[["source", "target"]].values.tolist(); 
edge2 = edge_df2[["source", "target"]].values.tolist();

for i in range(len(edge1)):
    G1.add_edge(edge1[i][0], edge1[i][1], weight = node_df[node_df["selector"] == edge1[i][0]][['fan_size_t1']].values[0][0])

for i in range(len(edge2)):
    G2.add_edge(edge2[i][0], edge2[i][1], weight = node_df[node_df["selector"] == edge2[i][0]][['fan_size_t2']].values[0][0])
```

To reproduce the graphs in the paper, only weakly connected nodes at T2 will be studied.

```python
def get_WCC(G, index):
    return G.subgraph(sorted(nx.weakly_connected_components(G), key=len, reverse=True)[index])
```

```python
# Remove from graphs
node_retain = list(get_WCC(G2, 0).nodes)
node_remove = [node for node in list(G2.nodes) if node not in node_retain]

G1.remove_nodes_from(node_remove); G2.remove_nodes_from(node_remove)
```

```python
# Edge color
def find_edge_color(edge_color):
    for i in range(len(edge_color)):
        edge = edge_color[i]
        if edge.startswith('a'):
            edge_color[i] = color[0]
        elif edge.startswith('n'):
            edge_color[i] = color[1]
        else: 
            edge_color[i] = color[2]
    return edge_color
edge_color1 = list(np.array([list(x) for x in list(G1.edges)])[:,0])
edge_color1 = find_edge_color(edge_color1)
edge_color2 = list(np.array([list(x) for x in list(G2.edges)])[:,0])
edge_color2 = find_edge_color(edge_color2)
```

### Create Link Growth Graph

```python
# Create graphs and add nodes
G_growth_a  = nx.DiGraph();
node_name = node_df[["selector"]].values.tolist(); node_name = sum(node_name, [])
node_pos = node_df[["x", "y"]].values.tolist(); node_pos = [tuple(x) for x in node_pos]
for i in range(len(node_name)):
    G_growth_a.add_node(node_name[i], pos = node_pos[i])
    
G_growth_n = G_growth_a.copy(); G_growth_p = G_growth_a.copy()
```

```python
# Add Edges
edgeincrease_df = edge_df[edge_df.T1 == 'FALSE']

edgeincrease_df_a = edgeincrease_df[edgeincrease_df.polarity == 'an#_vaccines']
edge = edgeincrease_df_a[["source", "target"]].values.tolist(); edge = [tuple(x) for x in edge]
G_growth_a.add_edges_from(edge)

edgeincrease_df_n = edgeincrease_df[edgeincrease_df.polarity == 'neutral']
edge = edgeincrease_df_n[["source", "target"]].values.tolist(); edge = [tuple(x) for x in edge]
G_growth_n.add_edges_from(edge)

edgeincrease_df_p = edgeincrease_df[edgeincrease_df.polarity == 'pro_vaccines']
edge = edgeincrease_df_p[["source", "target"]].values.tolist(); edge = [tuple(x) for x in edge]
G_growth_p.add_edges_from(edge)
```

```python
# Remove Outliers
G_growth_a.remove_nodes_from(node_remove); G_growth_n.remove_nodes_from(node_remove); G_growth_p.remove_nodes_from(node_remove)
```

```python
def find_proportion(edge_color):
    color = ['red', 'green', 'blue'];
    proportion = []
    for c in color:
        proportion.append(len(np.where(np.array(edge_color) == c)[0]))
    return proportion
```

```python
fig5 = plt.figure(figsize=(25,7))
plt.suptitle('Link Growth: Proportion of Targets', fontsize = 20)
mylabels = ["Anti-vaccination", "Neutral", "Pro-vaccination"]
edge_color = [edge_color_a, edge_color_n, edge_color_p]

def func(pct):
    return "{:1.1f}%".format(pct)

for i in range(3):
    plt.subplot(1,3,i+1)
    y = np.array(find_proportion(edge_color[i]))
    explode = np.zeros(3); explode[i] = 0.1
    color = ['tab:orange', 'tab:green', 'tab:blue']
    plt.pie(y, labels = mylabels, autopct=lambda pct: func(pct), shadow = True, explode = explode, textprops={'fontsize': 12}, colors = color)
    plt.title("From " + mylabels[i], fontsize = 18)
    
plt.show() 
fig5.savefig('Graphs/Proportion_Link_Growth.png', dpi = 300, transparent = False)
```
