# MA4M4 Extension of Bow-tie Decomposition in Political Blogs Network

Directed network with 1490 nodes and 33429 directed edges overall. 

Notice that the first 758 blogs are classified as left-leaning(liberal) and the remaining 732 are classified as right-leaning(conservative).

```python
import pandas as pd
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import betterspy
```

## A. Import Graph

```python
G = nx.DiGraph()
G.add_nodes_from(np.arange(1,1491))
```

```python
file = open('Data/data', 'r')
Lines = file.readlines()
 
count = 0; length = 0
for line in Lines:
    count += 1
    if count > 1 and count < 1491:
        tail = count-1
        linestr = line.strip(); lineint = linestr.split()
        for head in lineint:
            G.add_edge(tail, int(head))
```
