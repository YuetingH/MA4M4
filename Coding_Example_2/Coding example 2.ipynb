Try out the example in L3 slides. Changing the p from 0-1 inclusive and check how this influence the averge shortest path between vertices and average clustering coefficient. Plot the values of these two quantities normalized the first value. 

```python
p_list = np.arange(0,1,0.05)
N = 50;
k = 6;
average_sp = np.zeros((len(p_list),1))# average shortest path list
average_cc = np.zeros((len(p_list),1))# average clustering coef list
for idx,p in enumerate(p_list):

    # Generate Watts strogatz graph using the networkx random graph generator
    G_temp = nx.random_graphs.watts_strogatz_graph(N,k,p)
    # visualize a network using circular layout
    print('p value is ',p)
    nx.draw(G_temp,pos=nx.circular_layout(G_temp),with_labels=True,node_size=200,font_color='white')
    plt.show()
    average_sp[idx] = nx.average_shortest_path_length(G_temp)
    average_cc[idx] = nx.average_clustering(G_temp)
```

```python
plt.figure(figsize=(10,8))
# Normalize starting point to be 1
plt.plot(p_list,average_sp/(average_sp[0]),'o-',label = 'average shortest path')
plt.plot(p_list,average_cc/(average_cc[0]),'o-',label = 'average clustering coef')
plt.legend(fontsize=18)
plt.grid()
plt.ylabel('Normalized value',fontsize=18)
plt.xlabel(r'$p$',fontsize=18)
```

### A. ER Random Graph

```python
N = 100
p = 0.01
G_ER = nx.gnp_random_graph(N,p)
```

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
nx.draw(G_ER,
        pos=nx.spring_layout(G_ER,k=1.5,seed = 2022),
        with_labels=True, font_color='w')
plt.axis("on")
plt.subplot(1,2,2)
plt.spy(nx.to_numpy_matrix(G_ER))
```

```python
degree= [d for u,d in G_ER.degree()]
print("Average degree is", np.sum(degree)/N)
print("(N-1)*p is", (N-1)*p)
```

Let's compute (mean) average degree of large ensembles of graphs.

```python
SIM_NUM=1000;
mean_degree = np.zeros((SIM_NUM,1))
for i in range(0,SIM_NUM):
    G = nx.gnp_random_graph(N,p)
    degree = [d for u,d in G.degree()]
    mean_degree[i] = np.sum(degree)/N
print('average degree over all simulations',np.mean(mean_degree)) 
```

**Generate a relatively sparse network i.e p small, identify and visualize first and second connected components.**

```python
print(nx.number_connected_components(G_ER))
```

```python
gcc = G_ER.subgraph(sorted(nx.connected_components(G_ER), key=len, reverse=True)[0]);
second_cc =G_ER.subgraph(sorted(nx.connected_components(G_ER), key=len, reverse=True)[1])
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
nx.draw(G_ER,
        pos=nx.spring_layout(G_ER,k=1.5,seed = 2022),
        with_labels=True, font_color='w')
plt.axis("on");
plt.xlabel('Original graph');
plt.subplot(1,3,2)
nx.draw(gcc,
        pos = nx.spring_layout(gcc,k=1.5,seed = 2022),
        with_labels=True, font_color='w')
plt.axis("on");
plt.xlabel('largest connected component');
plt.subplot(1,3,3)
nx.draw(second_cc,
        pos = nx.spring_layout(second_cc,k=1.5,seed =2022),
        with_labels=True, font_color='w')
plt.axis("on");
plt.xlabel('Second large connected component');
```

**Vary c (average degree parameter) and see the emergence of giant component.**

```python
c_list = np.linspace(0,5,100)
N = 1000;
NUM_SIM = 5;
size_giant_component = np.zeros((len(c_list),NUM_SIM))
for idx, c in enumerate(c_list):
    for sim in range(NUM_SIM):
        # using c = (N-1)p
        G_temp = nx.gnp_random_graph(N,c/(N-1))
        largest_cc = max(nx.connected_components(G_temp), key=len)
        size_giant_component[idx,sim]=len(largest_cc)        
```

```python
plt.figure(figsize=(10,6))
plt.plot(c_list,np.mean(size_giant_component,axis=1),'b-',linewidth=2)
plt.plot([1,1],[0,N],'r--')
plt.ylim([0,N])
plt.ylabel('Average giant component size over the simulations',fontsize=12)
plt.xlabel(r'$c = (N-1)p$',fontsize=13)
plt.yticks(fontsize=18)
plt.xticks(fontsize=18);
```

### B.Configuration model
#### Get a degree sequence from Konect network which is ***undirected***.

```python
# Read the file using bulit in python commmand and built directly from edge_list
fh = open("out.moreno_zebra_zebra", "rb")
# fh = open("out.opsahl-openflights", "rb")
G_zebra= nx.read_edgelist(fh, comments='%',create_using=nx.Graph())
fh.close()
```

**Use the degree sequence and configuration model to generate network.**

```python
degree_sequence= [d for u,d in G_zebra.degree()]
G_cm = nx.configuration_model(degree_sequence)
```

**Visualize these two networks.**

```python
plt.figure(figsize=(20,8))
plt.subplot(1,2,1)
nx.draw(G_zebra,with_labels=True,pos=nx.spring_layout(G_zebra,k=1.5,seed=5),font_color='w')
plt.subplot(1,2,2)
nx.draw(G_cm,with_labels=True,pos=nx.spring_layout(G_cm,k=1.5,seed=5),font_color='w')
```
